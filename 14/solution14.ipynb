```python
from copy import deepcopy

def get_input(infile='input.txt'):
    with open(infile) as f:
        robots = []
        for line in f.readlines():
            p, v = line.strip().split()
            p = tuple(int(n) for n in p[2:].split(','))
            v = tuple(int(n) for n in v[2:].split(','))
            robots.append([p, v])
    w, h = (11, 7) if infile == 'input-small.txt' else (101, 103)
    return robots, w, h
```

```python
robots, w, h = get_input()
```

```python
def run_sim(rs, epochs=1):
    for i in range(epochs):
        for r in rs:
            px = r[0][0]
            py = r[0][1]
            dx = r[1][0]
            dy = r[1][1]
            r[0] = ((px + dx) % w, (py + dy) % h)
    return rs
```

```python
rs = deepcopy(robots)
rs = run_sim(rs, epochs=100)

mx = w // 2
my = h // 2
q = [0, 0, 0, 0]
for r in rs:
    px = r[0][0]
    py = r[0][1]
    if 0 <= px < mx:
        if 0 <= py < my:
            q[0] += 1
        elif my + 1 <= py:
            q[2] += 1
    elif mx + 1 <= px:
        if 0 <= py < my:
            q[1] += 1
        elif my + 1 <= py:
            q[3] += 1
f = 1
for qn in q:
    f *= qn
f
```

```python
import numpy as np

def print_board(rs, w, h):
    b = np.full((w, h), '.', dtype='U1')
    for r in rs:
        b[r[0]] = 'X'
    print("\n".join("".join(row) for row in b))
```

```python
rs = deepcopy(robots)

min_var_x = np.var([r[0][0] for r in rs])
min_var_y = np.var([r[0][1] for r in rs])

for i in range(w * h - 1):
    rs = run_sim(rs)
    var_x = np.var([r[0][0] for r in rs])
    var_y = np.var([r[0][1] for r in rs])
    if var_x < min_var_x and var_y < min_var_y:
        min_var_x = var_x
        min_var_y = var_y
        epoch = i + 1
print(epoch)
    
```

```python
print_board(run_sim(deepcopy(robots), epochs=7093), w, h)
```
